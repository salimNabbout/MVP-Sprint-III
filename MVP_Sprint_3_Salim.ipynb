```python
import pandas as pd

def ganhadores_por_ano(df):
    df['Ano'] = pd.to_datetime(df['Data_do_Sorteio'], format='%d/%m/%Y').dt.year
    ganhadores_por_ano = df.groupby('Ano').agg({
        'Ganhadores_6_acertos': 'sum',
        'Ganhadores_5_acertos': 'sum',
        'Ganhadores_4_acertos': 'sum'
    }).reset_index().sort_values('Ganhadores_6_acertos')
    return ganhadores_por_ano

ganhadores_ano = ganhadores_por_ano(df)
```

# **8-Quais s√£o os 6 numeros mais sorteados?**

```python
from collections import Counter

def numeros_mais_sorteados(df, n=6):
    numeros_sorteados = df[['Bola1', 'Bola2', 'Bola3', 'Bola4', 'Bola5', 'Bola6']]
    numeros_flat = numeros_sorteados.values.flatten()
    contador_numeros = Counter(numeros_flat)
    return contador_numeros.most_common(n)

mais_frequentes = numeros_mais_sorteados(df)
for numero, frequencia in mais_frequentes:
    print(f"N√∫mero {numero}: {frequencia} vezes")
```
